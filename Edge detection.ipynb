### Sobel Edge Detection :

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

def sobelOperator(img):
    container = np.copy(img)
    size = container.shape
    for i in range(1, size[0] - 1):
        for j in range(1, size[1] - 1):
            gx = (img[i - 1][j - 1] + 2*img[i][j - 1] + img[i + 1][j - 1]) - (img[i - 1][j + 1] + 2*img[i][j + 1] + img[i + 1][j + 1])
            gy = (img[i - 1][j - 1] + 2*img[i - 1][j] + img[i - 1][j + 1]) - (img[i + 1][j - 1] + 2*img[i + 1][j] + img[i + 1][j + 1])
            container[i][j] = min(255, np.sqrt(gx**2 + gy**2))
    return container
    pass
img1=cv2.imread("test2.jpg")
img = cv2.cvtColor(cv2.imread("test2.jpg"), cv2.COLOR_BGR2GRAY)
img = sobelOperator(img)
img = cv2.cvtColor(img, cv2.COLOR_GRAY2RGB)

# Create a figure with 2 subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Display the original image in the first subplot
ax1.imshow(img1)
ax1.set_title("Original Image")

# Display the image after the Sobel operator in the second subplot
ax2.imshow(img)
ax2.set_title("Image after Sobel Operator")

# Show the figure
plt.show()

```

### Canny Edge Detection:

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn')


loaded_image = cv2.imread('test.jfif') 
loaded_image = cv2.cvtColor(loaded_image,cv2.COLOR_BGR2RGB)


plt.figure(figsize=(10,10))
plt.imshow(loaded_image,cmap="gray")
plt.axis("off")
plt.show()

gray_image = cv2.cvtColor(loaded_image,cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(10,10))
plt.imshow(gray_image,cmap="gray")
plt.axis("off")
plt.show()
edged_image = cv2.Canny(gray_image, threshold1=30, threshold2=100)

plt.figure(figsize=(10,10))
plt.imshow(edged_image,cmap="gray")
plt.axis("off")
plt.show()
```
